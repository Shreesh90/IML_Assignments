```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('HousingPrice.csv')
```

```python
colA = np.array(data.price)
```

```python
colA = colA[:382]
colA
```

```python
colB = np.array(data.lotsize)[:382]

colC = np.array(data.bedrooms)[:382]

colD = np.array(data.bathrms)[:382]

len(colA)

mcolA = np.mat(colA)

mcolB = np.mat(colB)

mcolC = np.mat(colC)

mcolD = np.mat(colD)

xArray = []
yArray = []

m = len(colA) 

for i in range(0, m):
    xArray.append([1.0,float(colB[i]),float(colC[i]),float(colD[i])])
    yArray.append(float(colA[i]))
xmat = np.mat(xArray)
ymat = np.mat(yArray).T
```

```python
n = xmat.shape[1]
print(n)
theta=[]
for i in range(0,n):
    theta.append(1)
```

```python
theta
```

```python
def gradientDescent(x, y, theta, alpha, m, numIterations):
    m = x.shape[0]
    for j in range(0,numIterations):
        copy = theta.copy
        for i in range(0,m):
            rem.append( (theta*xmat[i] - y[i])*xmat[j][i])
        for i in range(0,m):
            copy[i] = copy[i] - (alpha/m)*rem[i]
        theta = copy
    return theta
```

```python
numIterations= 1000
alpha = 0.005
# theta = np.ones(n)

theta = gradientDescent(xmat, ymat, theta, alpha, m, numIterations)
print(theta)
```
